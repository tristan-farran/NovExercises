```python
import pandas as pd
```

```python
df = pd.read_csv('pokemon_data.csv')
```

```python
df[['Name', 'Type 1', 'Type 2', 'HP']][0:5] # extract columns with [['...']], and get the first 4 rows (kind of like array slicing)
```

```python
# iloc - integer-location lookup

# 0-indexed
# end parameter is excluded

df.iloc[3] # lookup row by row number (4th)
df.iloc[7:9] # lookup rows by row numbers (8th and 9th)
df.iloc[7, 3] # lookup element by coordinate (8th row, 4th column)
```

```python
for index, row in df.iterrows(): # iterate through rows (returns both index and row separately)
    print(row['Name'])

# though if you just wanted to get that column, df['Name'] would suffice
```

```python
# loc - label-based lookup,look up rows according to a conditional
df.loc[(df['Type 1'] == "Grass") & ((df['Type 2'] == "Poison") | (df['Type 2'] == "Fighting"))]
df.loc[~(df['Name'].str.contains('Mega') | (df['Name'].str.contains('%')))] # get rows which aren't "mega" or "%"
# note in pandas: and is represented by &, or is represented by |, not is represented by ~
```

```python
df.loc[df['Type 1'] == 'Fire', 'Type 2'] = 'Flamer' # conditional reassignment! Making type 2 for all fire type 1s "flamer"
df.loc[df['HP'] > 50, ['Generation','Legendary']] = ['Test 1', 'Test 2'] # multiple conditional reassignment setting rows with HP > 50 to have new Generation and Legendary Values
```

```python
df.describe() # get some general per row stats
```

```python
df.sort_values(['Type 1', 'HP'], ascending=[1,0]) # Sorts by Type 1 ascending first, and then by HP descending
```

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] # New column as the sum of the others
```

```python
df = df.drop(columns=['#']) # get rid of the # columm
```

```python
df['Total'] = df.iloc[:, 3:9].sum(axis=1) # sum the 4th-8th columns of every row horizontally
```

```python
cols = list(df.columns)
df = df[['Type 1','HP', 'Total', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']] # selecting and re-ordering columns, you could also do cols[0:4] + [cols[-1]] + cols[4:12] instead of the [names] but then if something changes the columns earlier things will break
```

```python
df.reset_index(drop=True, inplace=True) # if you removed rows and have jumps in your indices now and want to rest, you do it like this - drop = True means the old indices are not kept as a new column, as happens by default
```

```python
df.groupby (['Type 1']).mean().sort_values('Defense', ascending=False) # see the mean of the values for each Type 1 type, and sort by Defense to see the most defensive type
```

```python
df.groupby(['Type 1']).count() # groups by Type 1 and counts occurrences, replacing row values with the count

# groupby combined with other methods like count(), sum(), mean(), std() etc. is a useful way of getting subset statistics
```

```python
df['count'] = 1
df.groupby(['Type 1']).count()['count'] # counts occurrences of each type in a cleaner way, adding a count column to hold it, and then filtering for it after count
```

```python
df.to_csv('modified.csv', index=False) # export to CSV, without the indexes
```
